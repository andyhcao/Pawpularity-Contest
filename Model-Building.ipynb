```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
train_df = pd.read_csv('cleaned training data/clean_training_split')
train_df['scaled_transformed_score'] = scaler.fit_transform(train_df[['sqrt_tranformed score']])
train_df
```

```python
x = train_df.drop(['Id','sqrt_tranformed score','scaled_transformed_score','Unnamed: 0'],axis=1)
y = train_df['scaled_transformed_score']
x
```

```python
from sklearn.neighbors import RadiusNeighborsRegressor

rnr = RadiusNeighborsRegressor(radius = 1.50, n_jobs =-1, leaf_size=15)
rnr.fit(x,y)
```

```python
x_val = pd.read_csv('test.csv')
x_val = x_val.drop(['Id'],axis=1)
x_val
```

```python
predictions = pd.DataFrame(rnr.predict(x_val))
predictions
```

```python
pred = np.square(scaler.inverse_transform(predictions))
pred
```

# Linear Model

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
```

```python
train = pd.read_csv('cleaned training data/clean_training_split')
train
```

```python
x = train.drop(['Unnamed: 0', 'Id','sqrt_tranformed score'],axis=1)
y = train['sqrt_tranformed score']
```

```python
lm.fit(x,y)
```

```python
x_val = pd.read_csv('test.csv')

predictions = lm.predict(x_val.drop(['Id'], axis=1))
```

```python
pred= np.square(predictions)
pred
```

```python
submission = pd.DataFrame(pred, columns=['Pawpularity'])
sample_submission = pd.read_csv('sample_submission.csv')
sample_submission
```

```python
submission['Id'] = train['Id']
submission =submission[['Id','Pawpularity']]
submission
```
